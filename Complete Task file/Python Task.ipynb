```python
df = pd.read_excel('Data set (2).xlsx')
print("Null values per column:")
print(df.isnull().sum())
```

```python
print("\nTotal null values in dataset:", df.isnull().sum().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)

# For categorical columns â†’ fill with **mode**
for column in df.select_dtypes(include='object').columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
print("\nNull values after replacement:")
print(df.isnull().sum())
```

```python
# Calculate Measures of Central Tendency:
# 1. Mean
mean_values = df.mean()  # calculates mean for all numerical columns
print("Mean values for each numerical column:")
print(mean_values)
```

```python
# 2. Median
median_values = df.median()  # calculates median for all numerical columns
print("\nMedian values for each numerical column:")
print(median_values)
```

```python
# 3. Mode
mode_values = df.mode()  # calculates mode for all columns
print("\nMode values for each column:")
print(mode_values)
```
